```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("service_dataset.csv")
```

```python
# Drop identifiers and sensitive info
drop_cols = ['Number Plate','customer_id', 'name', 'email', 'mobile_number']
df.drop(columns=[col for col in drop_cols if col in df.columns], inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
# Convert date columns
date_cols = ['last_service_date', 'next_service_due_date', 'follow_up_date', 'feedback_date', 'last_call_date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Create numeric date difference features w.r.t. today's date
reference_date = pd.to_datetime('today').normalize()

if 'last_service_date' in df.columns:
    df['days_since_last_service'] = (reference_date - df['last_service_date']).dt.days

if 'next_service_due_date' in df.columns:
    df['days_until_next_service'] = (df['next_service_due_date'] - reference_date).dt.days

if 'follow_up_date' in df.columns:
    df['days_since_follow_up'] = (reference_date - df['follow_up_date']).dt.days

if 'feedback_date' in df.columns:
    df['days_since_feedback'] = (reference_date - df['feedback_date']).dt.days

if 'last_call_date' in df.columns:
    df['days_since_last_call'] = (reference_date - df['last_call_date']).dt.days
```

```python
df
```

```python
import pandas as pd

reference_date = pd.to_datetime('today').normalize()
max_service_period = 120
alert_threshold = 7

if 'next_service_due_date' in df.columns:
    # Convert to datetime if not done
    df['next_service_due_date'] = pd.to_datetime(df['next_service_due_date'], errors='coerce')
    
    # Calculate days until next service
    df['days_until_next_service'] = (df['next_service_due_date'] - reference_date).dt.days
    
    # Limit maximum period to 180 days
    df['days_until_next_service'] = df['days_until_next_service'].clip(upper=max_service_period)
    
    # Create alert flag (e.g., threshold = 7 days before due date)
    df['alert_due'] = df['days_until_next_service'].apply(
        lambda x: True if pd.notnull(x) and x <= alert_threshold else False)
    
    # Now you can filter all customers needing alert:
    alert_customers = df[df['alert_due'] == True]
    
    print(f"Total customers to alert: {len(alert_customers)}")
```

```python
# Drop original date columns
df.drop(columns=[col for col in date_cols if col in df.columns], axis = 1, inplace=True)
```

```python
def quanQual(df):
    quan=[]
    qual=[]
    for columnName in df.columns:
        #print(columnName)
        if(df[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual

quanQual(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Example: feedback vs. service booked
sns.boxplot(data=df, x='service_booked', y='feedback_score')
plt.title('feedback vs. service booked')
plt.show()
```
